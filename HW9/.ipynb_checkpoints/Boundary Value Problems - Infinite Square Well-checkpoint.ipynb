This curve oscillates about 0. Therefore, there will be a series of values of $E$ that yield $\psi_L(E) = 0$, which is what we expect from the analytical solutions. Let's look at the wave functions

This is close to what we want, but the value of the wave function at $x = L$ is not quite zero. We could continue modifying the value of $E$ until we get the desired result, but let's be more systematic about it and write a program that determines this energy for us

## Energy eigenvalues

The wave functions are not normalized, but they have the correct boundary conditions and appear to be given by $\sin(n\pi x/L)$

## Normalized Wave Functions

```python
from numpy import array, linspace, arange, pi, sqrt, sin
from pylab import plot, show, xlabel, ylabel, legend
%matplotlib inline
    
# Solve for wave function
def solvepsi(E):
    
    # Potential
    def V(x):
        return 0.0
    
    # Solve differential equation dr/dt = f(r,t)
    def f(r,x):
        psi = r[0]
        phi = r[1]
        fpsi = phi                              # dpsi/dx
        fphi = (2*m/hbar**2)*(V(x) - E)*psi     # dphi/dx
        return array([fpsi, fphi], float)
    
    r = array([0, 1.0], float) # r vector
    psiList = []
    
    for x in xvalues:
        
        # 4th order Runge-Kutta
        k1 = dx*f(r,x)
        k2 = dx*f(r + 0.5*k1, x + 0.5*dx)
        k3 = dx*f(r + 0.5*k2, x + 0.5*dx)
        k4 = dx*f(r + k3, x + dx)
        r += 1/6*(k1 + 2*k2 + 2*k3 + k4)
        psiList.append(r[0])
    return array(psiList)

# Determines the value of energy such that psiL(E) = 0 via secant method
def solve(E):
    E1 = E            # guess 1
    E2 = E1 + e       # guess 2
    target = e/1000   # desired accuracy
    
    # Secant method
    while abs(E1 - E2) > target:
        psi1 = solvepsi(E1)[-1]
        psi2 = solvepsi(E2)[-1]
        delta = -psi2*(E2 - E1)/(psi2 - psi1)
        E1 = E2
        E2 += delta
    
    return E2

# Normalizes the wave function via trapezoid rule
def normalize(psi):
    I = 0
    for i in range(len(psi) - 1):
        I += 0.5*dx*(psi[i]**2 + psi[i + 1]**2)
    
    return psi/sqrt(I)

# Constants
m = 9.1094e-31              # mass of electron (kg)
hbar = 1.0546e-34           # Planck's constant over 2*pi
e = 1.6022e-19              # electron charge (C)
L = 1e-10                   # Width of well (m)
N = 1000                    # number of points
dx = L/N                    # step size
xvalues = arange(0,L,dx)    # x values
Eguess = [6e-18, 4*6e-18, 9*6e-18] # guesses for energies of first three states

# Solve for wave functions and plot
plot(xvalues, 0*xvalues, "m--")
for n in range(len(Eguess)):
    E = solve(Eguess[n])
    Evals.append(E)
    psi = normalize(solvepsi(E))
    plot(xvalues, psi, "k.", markevery=20)
    plot(xvalues, sqrt(2/L)*sin((n+1)*pi*xvalues/L), "-", label="$\psi_{:d}$".format(n+1))
legend()
```

## Calculating Observables from $\psi(x)$

```python
from numpy import array, linspace, arange, pi, sqrt, sin
from pylab import plot, show, xlabel, ylabel, legend
%matplotlib inline
    
# Solve for wave function
def solvepsi(E):
    
    # Potential
    def V(x):
        return 0.0
    
    # Solve differential equation dr/dt = f(r,t)
    def f(r,x):
        psi = r[0]
        phi = r[1]
        fpsi = phi                              # dpsi/dx
        fphi = (2*m/hbar**2)*(V(x) - E)*psi     # dphi/dx
        return array([fpsi, fphi], float)
    
    r = array([0, 1.0], float) # r vector
    psiList = []
    
    for x in xvalues:
        
        # 4th order Runge-Kutta
        k1 = dx*f(r,x)
        k2 = dx*f(r + 0.5*k1, x + 0.5*dx)
        k3 = dx*f(r + 0.5*k2, x + 0.5*dx)
        k4 = dx*f(r + k3, x + dx)
        r += 1/6*(k1 + 2*k2 + 2*k3 + k4)
        psiList.append(r[0])
    return array(psiList)

# Determines the value of energy such that psiL(E) = 0 via secant method
def solve(E):
    E1 = E            # guess 1
    E2 = E1 + e       # guess 2
    target = e/1000   # desired accuracy
    
    # Secant method
    while abs(E1 - E2) > target:
        psi1 = solvepsi(E1)[-1]
        psi2 = solvepsi(E2)[-1]
        delta = -psi2*(E2 - E1)/(psi2 - psi1)
        E1 = E2
        E2 += delta
    
    return E2

# Normalizes the wave function via trapezoid rule
def normalize(psi):
    I = 0
    for i in range(len(psi) - 1):
        I += 0.5*dx*(psi[i]**2 + psi[i + 1]**2)
    
    return psi/sqrt(I)

# Calculates average position of particle
def average(x, psi):
    avg = 0
    for i in range(len(x) - 1):
        avg += 0.5*dx*(x[i]*psi[i]**2 + x[i + 1]*psi[i + 1]**2)
    return avg

# Constants
m = 9.1094e-31              # mass of electron (kg)
hbar = 1.0546e-34           # Planck's constant over 2*pi
e = 1.6022e-19              # electron charge (C)
L = 1e-10                   # Width of well (m)
N = 1000                    # number of points
dx = L/N                    # step size
xvalues = arange(0,L,dx)    # x values
Eguess = [6e-18, 4*6e-18, 9*6e-18] # guesses for energies of first three states

# Solve for wave functions and plot
for n in range(len(Eguess)):
    E = solve(Eguess[n])
    psi = normalize(solvepsi(E))
    plot(xvalues, psi, "k.", markevery=20)
    xavg = average(xvalues, psi)
    xsqavg = average(xvalues**2, psi)
    sigma = sqrt(xsqavg - xavg**2)
    print("n = {:d}: <x> = {:.2f}L, <x^2> = {:.2f}L^2, sigma = {:.2f}L".format(n+1, xavg/L, xsqavg/L**2, sigma/L))
    plot(xvalues, sqrt(2/L)*sin((n+1)*pi*xvalues/L), "-", label="$\psi_{:d}$".format(n+1))
legend()
show()
```
