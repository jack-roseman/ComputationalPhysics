## Electric Potential

```python
from numpy import zeros, max
from pylab import imshow, show, colorbar
%matplotlib inline

# Constants
M = 101           # Number of grid points on a side
VB = 0            # Voltage on box walls
VL = 1.0          # Voltage on left plate
VR = -1.0         # Voltage on right plate
target = 1e-5     # Target accuracy in volts
d = 30            # Plate separation in units of grid points
w = 61            # Plate width in units of grid points
xL = M//2 - d//2  # Position of left plate
xR = M//2 + d//2  # Position of right plate
yL = M//2 - w//2  # Bottom of left plate
yR = M//2 - w//2  # Bottom of right plate

# Arrays for potential
phi = zeros([M, M], float)
phi[yL:yL + w + 1, xL] = VL
phi[yR:yR + w + 1, xR] = VR
phiprime = zeros([M, M], float)

# Main loop
delta = 1.0
while delta > target:
    
    # Calculate new values of the potential
    for i in range(M):
        for j in range(M):
            
            # Apply boundary condition
            if i == 0 or i == M-1 or j == 0 or j == M-1:
                phiprime[i,j] = 0
            # Fixed potential for conductor
            elif (i >= yL and i <= yL + w and j == xL) or (i >= yR and i <= yR + w and j == xR):
                phiprime[i,j] = phi[i,j]
            # Relaxation method for Laplace equation
            else:
                phiprime[i,j] = (phi[i+1,j] + phi[i-1,j] + phi[i,j+1] + phi[i,j-1])/4
                
    # Calculate maximum difference from old values
    delta = max(abs(phi - phiprime))
    
    # Swap the two arrays
    phi,phiprime = phiprime,phi
    

# Plot results
imshow(phi, cmap="RdBu_r", origin="lower")
colorbar()
show()
```

## Electric Potential and Field

```python
from numpy import zeros, max, meshgrid, sqrt
from pylab import imshow, show, colorbar, quiver, streamplot
%matplotlib inline

# Constants
M = 101           # Number of grid points on a side
VB = 0            # Voltage on box walls
VL = 1.0          # Voltage on left plate
VR = -1.0         # Voltage on right plate
target = 1e-4     # Target accuracy in volts
d = 40            # Plate separation in units of grid points
w = 61            # Plate width in units of grid points
xL = M//2 - d//2  # Position of left plate
xR = M//2 + d//2  # Position of right plate
yL = M//2 - w//2  # Bottom of left plate
yR = M//2 - w//2  # Bottom of right plate

# Arrays for potential
phi = zeros([M, M], float)
phi[yL:yL + w + 1, xL] = VL
phi[yR:yR + w + 1, xR] = VR
phiprime = zeros([M, M], float)

# Arrays for electric field
Ex = zeros([M,M], float)
Ey = zeros([M,M], float)

# Main loop
delta = 1.0
while delta > target:
    
    # Calculate new values of the potential
    for i in range(M):
        for j in range(M):
            
            # Apply boundary condition
            if i == 0 or i == M-1 or j == 0 or j == M-1:
                phiprime[i,j] = 0
            # Fixed potential for conductor
            elif (i >= yL and i <= yL + w and j == xL) or (i >= yR and i <= yR + w and j == xR):
                phiprime[i,j] = phi[i,j]
            # Relaxation method for Laplace equation
            else:
                phiprime[i,j] = (phi[i+1,j] + phi[i-1,j] + phi[i,j+1] + phi[i,j-1])/4
                
    # Calculate maximum difference from old values
    delta = max(abs(phi - phiprime))
    
    # Swap the two arrays
    phi,phiprime = phiprime,phi
    
# calculate electric field
for i in range(M):
    for j in range(M):
        
        # calculate Ey
        if i == 0:
            Ey[i,j] = -(phi[i+1,j] - phi[i,j])        # forward difference
        elif i == M-1:
            Ey[i,j] = -(phi[i,j] - phi[i-1,j])        # backward difference
        else:
            Ey[i,j] = -(phi[i+1,j] - phi[i-1,j])/2    # central difference

        # calculate Ex
        if j == 0:
            Ex[i,j] = -(phi[i,j+1] - phi[i,j])        # forward difference
        elif j == M-1:
            Ex[i,j] = -(phi[i,j] - phi[i,j-1])        # backward difference
        else:
            Ex[i,j] = -(phi[i,j+1] - phi[i,j-1])/2    # central difference


# Plot results
X,Y = meshgrid(range(M), range(M))
stride = 5
x = X[::stride,::stride]
y = Y[::stride,::stride]
ex = Ex[::stride,::stride]
ey = Ey[::stride,::stride]
imshow(phi, extent=[0,M-1,0,M-1], cmap="RdBu_r", origin="lower")
colorbar()
quiver(x,y,ex,ey,pivot="middle", scale=10*max(sqrt(Ex**2 + Ey**2))) 
show()
```

```python
imshow(phi, extent=[0,M-1,0,M-1], cmap="RdBu_r")
colorbar()
numstream = 10
streampts = zeros([2*numstream, 2], float)
for n in range(numstream):
    xs = xL
    ys = yL + w/(numstream-1)*n
    streampts[n,0] = xs
    streampts[n,1] = ys
for n in range(numstream, 2*numstream):
    xs = xR
    ys = yR + w/(numstream-1)*(n - numstream)
    streampts[n,0] = xs
    streampts[n,1] = ys    
streamplot(X,Y,Ex,Ey, start_points=streampts, color="k")
show()
```
