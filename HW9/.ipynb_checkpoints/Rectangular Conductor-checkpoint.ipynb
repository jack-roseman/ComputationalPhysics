```python
from numpy import zeros, max, meshgrid, sqrt
from pylab import imshow, show, colorbar, quiver, streamplot
%matplotlib inline

# Constants
M = 101           # Number of grid points on a side
VB = 0            # Voltage on box walls
VC = 1
target = 1e-6     # Target accuracy in volts
Lx = 50
Ly = 50
x = M//2
y = M//2

# Arrays for potential
phi = zeros([M, M], float)
phi[(y - Ly//2):(y + Ly//2 + 1), (x - Lx//2):(x + Lx//2 + 1)] = VC
phiprime = zeros([M, M], float)


# Main loop
delta = 1.0
while delta > target:
    
    # Calculate new values of the potential
    for i in range(M):
        for j in range(M):
            
            # Apply boundary condition
            if i == 0 or i == M-1 or j == 0 or j == M-1:
                phiprime[i,j] = 0
            # Fixed potential for conductor
            elif (i >= y - Ly//2 and i <= y + Ly//2 and j >= x - Lx//2 and j <= x + Lx//2):
                phiprime[i,j] = phi[i,j]
            # Relaxation method for Laplace equation
            else:
                phiprime[i,j] = (phi[i+1,j] + phi[i-1,j] + phi[i,j+1] + phi[i,j-1])/4
                
    # Calculate maximum difference from old values
    delta = max(abs(phi - phiprime))
    
    # Swap the two arrays
    phi,phiprime = phiprime,phi
    

# Arrays for electric field
Ex = zeros([M,M], float)
Ey = zeros([M,M], float)


# calculate electric field
for i in range(M):
    for j in range(M):
        
        # calculate Ey
        if i == 0:
            Ey[i,j] = -(phi[i+1,j] - phi[i,j])        # forward difference
        elif i == M-1:
            Ey[i,j] = -(phi[i,j] - phi[i-1,j])        # backward difference
        else:
            Ey[i,j] = -(phi[i+1,j] - phi[i-1,j])/2    # central difference

        # calculate Ex
        if j == 0:
            Ex[i,j] = -(phi[i,j+1] - phi[i,j])        # forward difference
        elif j == M-1:
            Ex[i,j] = -(phi[i,j] - phi[i,j-1])        # backward difference
        else:
            Ex[i,j] = -(phi[i,j+1] - phi[i,j-1])/2    # central difference

# Plot results
X,Y = meshgrid(range(M), range(M))
stride = 5
x = X[::stride,::stride]
y = Y[::stride,::stride]
ex = Ex[::stride,::stride]
ey = Ey[::stride,::stride]
imshow(phi, extent=[0,M-1,0,M-1], cmap="RdBu_r", origin="lower")
colorbar()
quiver(x,y,ex,ey,pivot="middle", scale=10*max(sqrt(Ex**2 + Ey**2))) 
show()
```
