## Modified Infinite Square Well

```python
from numpy import array, linspace, arange, pi, sqrt, exp, log
from pylab import plot, show, xlabel, ylabel, legend
%matplotlib inline

# Potential
def V(x):
    a = 2*L/3
    w = L/10
    V0 = 10e-17
    return V0*exp(-(x-a)**2/(2*w**2))

from numpy import array, linspace, arange, pi, sqrt
from pylab import plot, show, xlabel, ylabel, legend
%matplotlib inline
    
# Solve for wave function
def solvepsi(E):
    def V(x):
        a = 2*L/3
        w = L/10
        V0 = 10e-17
        return V0*exp(-(x-a)**2/(2*w**2))

    # Solve differential equation dr/dt = f(r,t)
    def f(r,x):
        psi = r[0]
        phi = r[1]
        fpsi = phi                              # dpsi/dx
        fphi = (2*m/hbar**2)*(V(x) - E)*psi     # dphi/dx
        return array([fpsi, fphi], float)
    
    r = array([0, 1.0], float) # r vector
    psiList = []
    
    for x in xvalues:
        
        # 4th order Runge-Kutta
        k1 = dx*f(r,x)
        k2 = dx*f(r + 0.5*k1, x + 0.5*dx)
        k3 = dx*f(r + 0.5*k2, x + 0.5*dx)
        k4 = dx*f(r + k3, x + dx)
        r += 1/6*(k1 + 2*k2 + 2*k3 + k4)
        psiList.append(r[0])
    return array(psiList)

# Determines the value of energy such that psiL(E) = 0 via secant method
def solve(E):
    E1 = E            # guess 1
    E2 = E1 + e       # guess 2
    target = e/1000   # desired accuracy
    
    # Secant method
    while abs(E1 - E2) > target:
        psi1 = solvepsi(E1)[-1]
        psi2 = solvepsi(E2)[-1]
        delta = -psi2*(E2 - E1)/(psi2 - psi1)
        E1 = E2
        E2 += delta
    
    return E2

# Constants
m = 9.1094e-31              # mass of electron (kg)
hbar = 1.0546e-34           # Planck's constant over 2*pi
e = 1.6022e-19              # electron charge (C)
L = 1e-10                   # Width of well (m)
N = 1000                    # number of points
dx = L/N                    # step size
xvalues = arange(0,L,dx)    # x values
```

## Plot of $V(x)$

```python
potentialVals = [V(x) for x in xvalues]
plot(xvalues, potentialVals)
xlabel("x")
ylabel("Potential")
show()
```

## Plot of $\psi_L(x)$ vs. $E$

```python
# Make a plot of psiL(E) vs. E
Evals = linspace(0, 1000*e, 100) ##J
psiLvalues = []
for E in Evals:
    psiLvalues.append(solvepsi(E)[-1])

plot(Evals, psiLvalues)
plot(Evals, 0*Evals, "c--")
xlabel("$E$ (J)")
ylabel("$\psi_L$")
show()
```

## Energy Levels

```python
Eguess = [5e-18, 7e-17, 12e-17] # guesses for energies of first three states
E1 = solve(Eguess[0])
E2 = solve(Eguess[1])
E3 = solve(Eguess[2])
psi1 = solvepsi(E1)
psi2 = solvepsi(E2)
psi3 = solvepsi(E3)
print("Guesses:", E1, E2, E3)
```

## Normalized Wave Functions

```python
# Plot for first three normalized wave functions
def normalize(psi):
    I = 0
    for i in range(len(psi) - 1):
        I += 0.5*dx*(psi[i]**2 + psi[i + 1]**2)
    return psi/sqrt(I)

plot(xvalues, 0*xvalues, "m--")
plot(xvalues, normalize(psi1), label="$\psi_1$")
plot(xvalues, normalize(psi2), label="$\psi_2$")
plot(xvalues, normalize(psi3), label="$\psi_3$")
legend()
show()

plot(xvalues, 0*xvalues, "m--")
plot(xvalues, normalize(psi1)**2, label="$\psi_1$")
plot(xvalues, normalize(psi2)**2, label="$\psi_2$")
plot(xvalues, normalize(psi3)**2, label="$\psi_3$")
legend()
show()
```

## Probability of Occupying Classically Forbidden Region

```python
xturnleft = a - sqrt(2*w**2*log(V0/E1))
xturnright = a + sqrt(2*w**2*log(V0/E1))

def probability(x1, x2, psi):
    lastindex = int(x2/dx)
    out = 0.0
    xleft = x1
    xright = x1 + dx
    while xright< x2:
        out = 0.5*dx*psi[int(xright/dx)]**2
        xleft = xright
        xright = xleft + dx
    return out

print(probability(xturnleft, xturnright, normalize(psi1))) ##nonzero probability, so electron can tunnel
```
