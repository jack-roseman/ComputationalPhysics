```python
##Question 1
from math import sqrt

def primesupto(k):
    primelist = [2]
    for n in range(3, k):
        hasprimefactor = False
        for i in primelist:
            if i > sqrt(n):
                break
            if n % i == 0:
                hasprimefactor = True
                break
        if not hasprimefactor:
            primelist.append(n)
    return primelist
primes = primesupto(1000000)
print("There are {} primes between 2 and 1,000,000".format(len(primes)))
```

```python
##Question 2
from pylab import plot, show
import numpy as np
file1 = np.loadtxt("sunspots.txt")

# rows = file1.read().splitlines()
tvals = file1[:,0].tolist()
yvals = file1[:,1].tolist()

print("Before Smoothing:")
plot(tvals, yvals)
show()

smoothtvals = []
smoothyvals = []
r = int(input("Enter smoothing factor:"))

for i in range(r, 1000 - r):
    sum = 0
    for j in range(i - r, i + r):
        sum += yvals[j]
    smoothtvals.append(tvals[i])
    smoothyvals.append(sum / (2 * r + 1))


print("After Smoothing:")
plot(smoothtvals, smoothyvals)
show()
```

```python
# question 3
%matplotlib inline
from matplotlib import pyplot as plt 

a1 = 15.8 #MeV
a2 = 18.3 #MeV
a3 = 0.714 #MeV
a4 = 23.2 #MeV
a5 = 0.0

massNumbers = []
bindingEnergies = []
for Z in range(1, 100):
    bestA = Z
    maxBindingEnergyPerNucleon = 0
    for A in range(Z, 3*Z):
        if (not A % 2 == 0):
            a5 = 0.0
        elif (A % 2 == 0 and not Z % 2 == 0):
            a5 = -12.0
        B = a1*A - a2*A**(2/3) - a3*Z**2/A**(1/3) - a4*(A - 2*Z)**2/A + a5/A**(1/2)
        if B/A > maxBindingEnergyPerNucleon:
            maxBindingEnergyPerNucleon = B/A
            bestA = A
    massNumbers.append(bestA)
    bindingEnergies.append(maxBindingEnergyPerNucleon)


print("d) Maximum binding energy per Nucleon occurs at Atomic Number:", bindingEnergies.index(max(bindingEnergies)), "which is Nickel (Mass Number: 58)")
plt.scatter(massNumbers, bindingEnergies)
plt.xlabel("Mass Number", fontsize=12)
plt.ylabel("Binding Energy / Nucleon (eV)", fontsize=12)
show()
```
