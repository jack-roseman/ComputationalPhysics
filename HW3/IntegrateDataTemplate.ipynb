```python
# Enter code here that calculates and plots the acceleration, velocity and position as functions of time
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 1000
raw = np.loadtxt("forces.txt")

points = list(zip(*raw))
xvals = points[0]
tmp = points[1]
yvals = []
for i in range(len(tmp)):
    yvals.append(tmp[i] / 1.30)

plt.title("acceleration vs time")
plt.xlabel("time (s)")
plt.ylabel("acceleration (m/s^2)")
plt.plot(xvals, yvals)
plt.show()

vels = [0]
times = [0]
for i in range(1, len(xvals)):
    x1 = xvals[i-1]
    y1 = yvals[i-1]
    x2 = xvals[i]
    y2 = yvals[i]
    dv = 0.5*(y1+y2)*(x2-x1)
    vels.append(vels[i-1]+dv)
    times.append(0.5*(x1+x2))

plt.title("velocity vs time")
plt.xlabel("time (s)")
plt.ylabel("velocity (m/s)")
plt.plot(times, vels)
plt.show()

pos = [0]
times2 = [0]
for i in range(1, len(times)):
    x1 = times[i-1]
    y1 = vels[i-1]
    x2 = times[i]
    y2 = vels[i]
    dx = 0.5*(y1+y2)*(x2-x1)
    pos.append(pos[i-1]+dx)
    times2.append(0.5*(x1+x2))

plt.title("displacement vs time")
plt.xlabel("time (s)")
plt.ylabel("displacement (m)")
plt.plot(times2, pos)
plt.show()
```
