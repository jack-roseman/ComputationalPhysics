```python
# Enter code that calculates and plots the gravitational field
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 100

def f(x, y, z):
    if x == 0.0 and y == 0.0 and z == 0.0:
        return float('inf')
    return 1.0 / (x**2 + y**2 + z**2)**1.5
    
def integrate(z):
    a = -0.5
    b = 0.5
    c = -0.5
    d = 0.5
    dx = (b-a)/N
    dy = (d-c)/N
    s = 0.25*(f(a, c, z) + f(b, d, z))
    for i in range(1, N):
        for j in range(1, N):
            s += f(a + i*dx, c + j*dy, z)
    return dx*dy*s

def g(z):
    return z*integrate(z)

xvals = np.linspace(0, 1, 100)
plt.plot(xvals, [g(z) for z in xvals])
plt.xlabel("z")
plt.ylabel("g")
plt.show()
```

```python
#I got rid of artifict by setting the delta z of the calculation to a number equal to (b-a) / N. This way, every corresponding sample point z maps perfectly to a computed value for the sample points used to compute over the x and y intervals
```
