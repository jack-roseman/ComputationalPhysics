## Comparison between 4th-Order Runge-Kutta and Verlet Methods

```python
# Author: Bob Johnson robertjo@physics.upenn.edu

from numpy import array, sin, cos, pi, arange
from pylab import plot, show, legend, xlabel, ylabel, ylim
%matplotlib inline

# this program solves dr/dt = f(r,t)
def alpha(theta):
    return -g/l*sin(theta)

def f(r,t):
    theta = r[0]
    omega = r[1]
    ftheta = omega
    fomega = alpha(theta)
    return array([ftheta, fomega], float)

# parameters
m = 1.0                      # mass of pendulum bob
g = 1.0                      # gravitational acceleration
l = 1.0                      # length of pendulum
dt = 0.05                    # step size
t0 = 0                       # initial time
tf = 100.0                   # final time
tpoints = arange(t0, tf, dt) # array of time values

# Initial conditions
theta = pi/3                                    # initial angle
omega = 0.0                                     # initial angular velocity
rRK = array([theta, omega])                     # initial conditions
E0 = 0.5*m*l**2*omega**2  + m*g*l*(1-cos(theta))# initial total mechanical energy 

# Data lists
thetaRK = []
omegaRK = []
thetaV = []
omegaV = []

# Perform half step forward for Verlet method
omegamid = omega + alpha(theta)*0.5*dt

for t in tpoints:
    
    # Save data
    thetaRK.append(rRK[0])
    omegaRK.append(rRK[1])
    thetaV.append(theta)
    omegaV.append(omega)
    
    #4th order Runge-Kutta
    k1 = dt*f(rRK, t)
    k2 = dt*f(rRK + 0.5*k1, t + 0.5*dt)
    k3 = dt*f(rRK + 0.5*k2, t + 0.5*dt)
    k4 = dt*f(rRK + k3, t + dt)
    rRK += 1/6*(k1 + 2*k2 + 2*k3 + k4)
    
    # Verlet method
    theta += omegamid*dt
    omega = omegamid + alpha(theta)*0.5*dt
    omegamid += alpha(theta)*dt
    
    t += dt

# Energy arrays
ERK = 0.5*m*l**2*array(omegaRK)**2 + m*g*l*(1-cos(array(thetaRK))) # RK4 Energy
EV  = 0.5*m*l**2*array(omegaV)**2  + m*g*l*(1-cos(array(thetaV)))  # Verlet Energy
    
# Plot results
plot(tpoints, thetaRK, "r-", label="RK4")
plot(tpoints, thetaV, "b-", label="Verlet")
xlabel("$t$ (s)")
ylabel("$\\theta$ (rad)")
legend()
show()

plot(tpoints, omegaRK, "r-", label="RK4")
plot(tpoints, omegaV, "b-", label="Verlet")
xlabel("$t$ (s)")
ylabel("$\omega$ (rad/s)")
legend()
show()

plot(tpoints, ERK, "r-", label="RK4")
plot(tpoints, EV, "b-", label="Verlet")
xlabel("$t$ (s)")
ylabel("$E$ (J)")
ylim(0.999999*E0, E0)
legend()
show()
```

We can see the the RK4 method is losing energy. Even though the energy in the Verlet method fluctuates, the energy returns to the same value at each oscillation, thereby conserving energy in the long run

## Time-Reversal of Methods

```python
# Author: Bob Johnson robertjo@physics.upenn.edu

from numpy import array, sin, pi


# this program solves dr/dt = f(r,t)
def alpha(theta):
    return -g/l*sin(theta)

def f(r,t):
    theta = r[0]
    omega = r[1]
    ftheta = omega
    fomega = alpha(theta)
    return array([ftheta, fomega], float)

# parameters
m = 1.0                   # mass of pendulum bob
g = 1.0                   # gravitational acceleration
l = 1.0                   # length of pendulum
dt = 0.05                 # step size

# Initial conditions
t = 0                     # initial time
theta = pi/3              # initial angle
omega = 0.0               # initial angular velocity
r = array([theta, omega]) # initial conditions

print("{:20s} Theta0 = {:.16f}, Omega0 = {:.16f}".format("Initial Conditions:", theta, omega))

#4th order Runge-Kutta forward step
k1 = dt*f(r, t)
k2 = dt*f(r + 0.5*k1, t + 0.5*dt)
k3 = dt*f(r + 0.5*k2, t + 0.5*dt)
k4 = dt*f(r + k3, t + dt)
r += 1/6*(k1 + 2*k2 + 2*k3 + k4)
t += dt

#4th order Runge-Kutta backward step
k1 = -dt*f(r, t)
k2 = -dt*f(r + 0.5*k1, t - 0.5*dt)
k3 = -dt*f(r + 0.5*k2, t - 0.5*dt)
k4 = -dt*f(r + k3, t - dt)
r += 1/6*(k1 + 2*k2 + 2*k3 + k4)
t += -dt

print("{:20s} Theta0 = {:.16f}, Omega0 = {:.16f}".format("RK4:", r[0], r[1]))

# Verlet method forward step
omegamid = omega + alpha(theta)*0.5*dt
theta = theta + omegamid*dt
omega = omegamid + alpha(theta)*0.5*dt

# Verlet method backward step
omegamid = omega - alpha(theta)*0.5*dt
theta = theta - omegamid*dt
omega = omegamid - alpha(theta)*0.5*dt

print("{:20s} Theta0 = {:.16f}, Omega0 = {:.16f}".format("Verlet:", theta, omega))
```
