## (a) Orbit of the Moon

```python
from numpy import sqrt, arange, array, add
from pylab import plot, xlabel, ylabel, show, title, legend

import time
start = time.time()

def a(r):
    d = sqrt(r[0]**2 + r[1]**2)
    return -G*M*r/d**3

t = 0
dt = 3600 #1 hour in seconds
M = 5.972e24  #kg
G = 6.67408e-11  #Nm^2/kg^2
x0 = 3.626e8  #m
y0 = 0.0 #m
vx0 = 0.0 #m/s
vy0 = 1.077e3 #m/s
xpoints = [x0]
ypoints = [y0]
rpoints = [sqrt(x0**2 + y0**2)] ##magnitude of distance vector
vpoints = [sqrt(vx0**2 + vy0**2)] ##magnitude of velocity vector
tpoints = [t]

r = array([x0, y0], float)
v = array([vx0, vy0], float)

# Perform half step forward for Verlet method
vmid = v + a(r)*0.5*dt

norbit = 0

while norbit < 1:
    yold = r[1]
    # Verlet method
    r = r + vmid*dt
    v = vmid + a(r)*0.5*dt
    vmid = vmid + a(r)*dt
    t += dt
    
    xpoints.append(r[0])
    ypoints.append(r[1])
    rpoints.append(sqrt(r[0]**2 + r[1]**2)) ##magnitude of distance vector
    vpoints.append(sqrt(v[0]**2 + v[1]**2)) ##magnitude of velocity vector
    tpoints.append(t)
    
    ynew = r[1]
    if yold < 0 and ynew > 0:
        norbit += 1

print(time.time() - start, "seconds")
# plot results
plot(0, 0, "g.") #greet dot to indicate the position of earth
plot(xpoints, ypoints, ".")
title("1 orbit")
xlabel("x (m)")
ylabel("y (m)")
show()
```
